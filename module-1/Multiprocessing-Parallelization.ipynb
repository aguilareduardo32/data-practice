```python
import multiprocess
```

```python
def square(x):
    return x*x

data = [x for x in range(100)]
```

```python
%%time

seq = [square(x) for x in data]
print(seq)
```

```python
%%time

pool = multiprocessing.Pool()
result = pool.map(square, data)
pool.terminate()
pool.join()
print(result)
```

```python
def find_markdowns(dirt):
    files = os.listdir(dirt)
    
    markdowns = [dirt + '/' + file for file in files if file.endswith('.md')]
    return markdowns
```

```python
%%time 

result = [find_markdowns(dirt) for dirt in dirs]
flatten = [file for sublist in result for file in sublist]
flatten
```
