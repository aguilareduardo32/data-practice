```python
data['Fatal (Y/N)'] =data['Fatal (Y/N)'].astype(str)
```

```python
### age less than 10 from ten -20 , 20 -40, 40 -60, 60-100
## If boat or bouting, bouting
## Fatal = Y, N, unknow 
#Species remove space from title column. 
```

```python
def fatal_func(x):
    
    if x == ' N' or  x == 'N ' or x == 'n':
        return 'N'
    elif x == 'F':
        return 'Y'
    elif x =='#VALUE!' or x == 'UNKNOWN' or x == 'nan':
        
        return 'UNKNOWN' 
    elif x == 'Y' or x == 'N':
        return x
```

```python
data['Fatal (Y/N)1'] = data['Fatal (Y/N)'].apply(fatal_func)
```

```python
data['Fatal (Y/N)1'].value_counts()
```
