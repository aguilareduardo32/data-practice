```python
#How does the autoregresive model predicts outcomes, using the same model and diferent data?
```

```python
import pandas as pd

occupancy = pd.read_csv(r'C:\Users\Yael Aguilar\Downloads\occupancy.csv')
occupancy.head()
```

```python
occupancy.date = pd.to_datetime(occupancy.date)
occupancy.index = pd.DatetimeIndex(occupancy.date, freq='H')
occupancy.index
```

```python
from statsmodels.tsa.ar_model import AR

train = occupancy.CO2[:-10]
model = AR(occupancy.CO2[:-10])
model_fit = model.fit(maxlag=1)
predictions = model_fit.predict(start=len(train), end=len(train)+ 9, dynamic = False)
```

```python
pd.DataFrame({'reals': occupancy.CO2[-10:],'predicted': predictions})
```

```python
from statsmodels.tsa.arima_model import ARMA

model = ARMA(occupancy.CO2[:-3], order =(0,1))
model_fit = model.fit(disp=False)
predictions = model_fit.predict(len(occupancy.CO2)-3, len(occupancy.CO2)-1)




```

```python
pd.DataFrame({'observed': occupancy.CO2[-3:], 'predicted': predictions})
```

```python
from statsmodels.tsa.arima_model import ARIMA

model = ARMA(occupancy.CO2[:-3], order =(0,1))
model_fit = model.fit(disp=False)
predictions = model_fit.predict(len(occupancy.CO2)-3, len(occupancy.CO2)-1)


pd.DataFrame({'observed': occupancy.CO2[-3:], 'predicted': predictions})
```

```python
model = ARMA(occupancy.CO2, order=(2,1))
model_fit=model.fit(disp=False)
predictions = model_fit.predict(len(occupancy.CO2)-3, len(occupancy.CO2)-1)
```
