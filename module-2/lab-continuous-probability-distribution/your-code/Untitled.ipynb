```python
import numpy as np
from scipy.stats import uniform 
import matplotlib.pyplot as plt 



#Standard uniform distribution
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,1, 100)
ax.plot(x, uniform.pdf(x))
```

```python
#UNIFROM DISTRIBUTION BETWEEN POINT A, B

fig, ax = plt.subplots(1, 1)
a= 1
b= 3
x = np.linspace(a, b, 100)
y = uniform.pdf(x, a, b)
ax.plot(x, y)
```

```python
import math 
from scipy.stats import norm 
normal = norm(0, math.sqrt(9))
print(normal.pdf(4))
print(normal.cdf(2))
print(normal.rvs())
fig, ax = plt.subplots(1, 1)
x = np.linspace(-3, 3, 1000)
y = norm.pdf(x)
ax.plot(x, y)
plt.show()
```

```python
#Importing required libraries
from scipy.stats import uniform
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
 
#taking random variables from Uniform distribution
data = uniform.rvs(size = 100000, loc = 1, scale=5)
 
#Plotting the results
sb.set_style('whitegrid')
ax = sb.distplot(data, bins = 6, color = 'k')
ax.set(xlabel = 'interval')
plt.show()
```
