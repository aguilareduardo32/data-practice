```python
conditions = [students.absences <=4, (students.absences >=5) & (students.absences<=8), students.absences > 8]
choices = ['low', 'medium', 'high']
students['absencesmean2'] = np.select(conditions, choices)
pd.crosstab(index=students.traveltime, columns = students.absencesmean2)
```

```python
students['absencesmean'] = np.where(students.absences<= 4, 'low',
                                   (np.where(students.absences <8, 'medium', 'high')))
pd.crosstab(index =students.traveltime, columns=  students.absencesmean)
```
